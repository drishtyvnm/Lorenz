The MIT metereologists **Edward Lorenz** constructed a highly simplified model of a *convecting fluid*. This simple model also displays a wide variety of behavior and for some parameter values is chaotic. The system is given by $ \dot{x} = σ(y-x), \dot{y}=rx-y-xz, \dot{z}=xy-bz$, where $x$ maesures the rate of concetive overtuning, $y$ and $z $measures the horizontal and vertical temperature variation,
$σ$ is the *Prandtl number*, $r$ is *the Rayleigh numner* and $b$
 is the *scaling factor*. Thus, the surprising conclusion derived by Lorenz is now popularly known as *butterfly effect.*



```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
sigma, beta, rho = 10, 2.667, 28
x0, y0, z0 = 0, 1, 1.05
```

```python
tmax, n = 100, 1000

```

```python
def Lorenz(X, t, sigma, beta, rho):
  """The Lorenz equations."""
  x, y, z = X
  dx = -sigma * (x - y)
  dy = rho * x - y - x * z
  dz = -beta * z + x * y
  return dx, dy, dz
```

```python
t = np.linspace(0, tmax, n)
f = odeint(Lorenz, (x0, y0, z0,), t, args=(sigma, beta, rho))
x, y, z = f.T
```

```python
fig=plt.figure()
ax = Axes3D(fig)
ax.plot(x, y, z, 'b-' , lw=.5)
ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('y', fontsize=15)
ax.set_zlabel('z', fontsize=15)
plt.tick_params(labelsize=15)
ax.set_title('Lorenz Attractor' , fontsize=15)
plt.show()
```
